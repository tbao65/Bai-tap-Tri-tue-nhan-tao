```python
import osmnx as ox
import networkx as nx
```

```python
G = ox.graph_from_place('Quận 10, Hồ Chí Minh, Việt Nam', network_type='drive')
origin = (10.76172, 106.66821)
destination = (10.77324, 106.65792)
origin_node = ox.distance.nearest_nodes(G, X=origin[1], Y=origin[0], return_dist=False)
destination_node = ox.distance.nearest_nodes(G, X=destination[1], Y=destination[0], return_dist=False)
```

```python
route1 = nx.shortest_path(G, source=origin_node, target=destination_node, weight="length", method='dijkstra')
route2 = nx.astar_path(G, source=origin_node, target=destination_node, heuristic=None, weight="length", cutoff=None)
```

```python
fig, ax = ox.plot_graph_routes(
    G,
    [route1, route2],
    bgcolor='white',
    route_colors=['red', 'blue'],
    route_linewidths=[8,4],
    node_color='black'
)
```

**Thuật toán Dijkstra**

Ưu điểm:
*   Đảm bảo độ chính xác tuyệt đối
*   Dễ triển khai, Không cần phải thiết lập hay tính toán thêm các hàm ước lượng phức tạp.

Nhược điểm:
*   Chậm và lãng phí tài nguyên
*   Không phù hợp cho bản đồ lớn
