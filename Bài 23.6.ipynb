```python
import osmnx as ox
```

```python
G = ox.graph_from_place('Quận 10, Hồ Chí Minh, Việt Nam', network_type='drive')
stats = ox.basic_stats(G)
print(f"Tổng số nút giao: {stats['n']}")
print(f"Tổng số đoạn đường: {stats['m']}")
print(f"Tổng chiều dài các đoạn đường: {stats['edge_length_total']:.2f}")
print(f"Chiều dài trung bình mỗi đoạn đường: {stats['edge_length_avg']:.2f}")
print(f"Tổng số giao lộ: {stats['intersection_count']}")
```

```python
fig, ax = ox.plot_graph(G)
```
