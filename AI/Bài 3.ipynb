```python
import folium
from folium.plugins import HeatMap
import numpy as np
import pandas as pd
```

```python
n = 1000
latitude_min, latitude_max = 10.7, 10.9
longitude_min, longitude_max = 106.6, 106.8
dataset = {
    "latitude": np.random.uniform(latitude_min, latitude_max, n),
    "longitude": np.random.uniform(longitude_min, longitude_max, n),
    "orders": np.random.randint(1, 20, n)
}
data = pd.DataFrame(dataset)
data.head()
```

```python
m = folium.Map([10.827850, 106.863536], zoom_start=11)
HeatMap(data[['latitude', 'longitude', 'orders']].values.tolist()).add_to(m)
m
```

1. Các vùng có màu nóng (nhiệt cao) là nơi có mật độ đơn hàng lớn
- Đây là khu vực cầu cao (có thể là khu đông dân cư, khu thương mại lớn...)
- Khuyến nghị nên tăng số lượng nhân viên giao hàng và xem xét đặt kho ở quanh khu vực này.
2. Các vùng có màu lạnh là nơi có mật độ đơn hàng nhỏ
- Dân cư ở khu vực này có nhu cầu mua sắm thấp hoặc dịch vụ chưa được khai thác hiệu quả.
- Nên xem xét kỹ trước khi ra quyết định đầu tư, có thể chạy thử một số chiến dịch marketing để khảo sát nhu cầu.
