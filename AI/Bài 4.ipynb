```python
import geopandas as gpd
import pandas as pd
import numpy as np
import plotly.express as px
import json
```

```python
vn_map = gpd.read_file("Provinces.geojson")
```

```python
vn_map['Doanh_Thu'] = np.random.randint(500, 50000, size=len(vn_map))
geojson_data = json.loads(vn_map.to_json())
```

```python
fig = px.choropleth(
    vn_map,
    geojson=geojson_data,
    locations="TinhThanh",
    featureidkey="properties.TinhThanh",
    color="Doanh_Thu",
    color_continuous_scale="bluyl"
)
fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

- Các tỉnh có màu đậm là những khu vực có doanh thu cao, có thể cân nhắc gia tăng đầu tư để kiếm thêm lợi nhuận.
- Các tỉnh có màu nhạt là những khu vực có doanh thu thấp, nên xem xét lại chiến lược bán hàng/ đầu tư.
