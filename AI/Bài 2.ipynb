```python
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
from geopy.distance import distance
import folium
import time
```

```python
geolocator = Nominatim(user_agent="geoapi")
center = "Independence Palace, Ho Chi Minh City"
center_location = geolocator.geocode(center)
center_coordinate = (center_location.latitude, center_location.longitude)
print(f"Tên: {center_location.address}")
print(f"Tọa độ: {center_location.latitude}, {center_location.longitude}")

m = folium.Map(location=center_coordinate, zoom_start=12)
folium.Marker(center_coordinate, tooltip=f"Điểm trung tâm: {center}", icon=folium.Icon(icon="star")).add_to(m)
```

```python
cityscapes = [
    "Saigon Central Post Office, Ho Chi Minh City",
    "Ho Chi Minh City Hall, Ho Chi Minh City",
    "Ben Thanh Market, Ho Chi Minh City",
    "Nhà thờ Đức Bà, Ho Chi Minh City",
    "Phố đi bộ Nguyễn Huệ, Ho Chi Minh City",
    "Saigon Zoo and Botanical Gardens, Ho Chi Minh City",
    "War Remnants Museum, Ho Chi Minh City",
    "Landmark 81, Ho Chi Minh City",
    "Nhà hát Thành phố Hồ Chí Minh, Ho Chi Minh City",
    "Tao Dan Park, Ho Chi Minh City"
]
```

```python
for cityscape in cityscapes:
  location = geolocator.geocode(cityscape)
  if location:
    coordinate = (location.latitude, location.longitude)
    distance = geodesic(center_coordinate, coordinate).km
    folium.Marker(coordinate, popup=f"Khoảng cách đến trung tâm: {distance:.2f}km", tooltip=f"{location}").add_to(m)
    folium.PolyLine([center_coordinate,coordinate], color="blue").add_to(m)
    time.sleep(1)
```
