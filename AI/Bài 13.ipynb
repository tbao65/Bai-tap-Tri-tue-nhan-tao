```python
import folium
```

```python
m = folium.Map(location=[10.738833, 106.702258], zoom_start=13)
layer_vung = folium.FeatureGroup(name='Bán kính phục vụ')
layer_diem = folium.FeatureGroup(name='Vị trí cửa hàng')
layer_tuyen = folium.FeatureGroup(name='Tuyến đường vận chuyển')
```

```python
folium.Circle(
    location=[10.74659, 106.69788],
    radius=1000, color='red', fill=True, fill_opacity=0.2,
    tooltip='Khu vực kho hàng'
).add_to(layer_vung)

folium.Circle(
    location=[10.72964, 106.71891],
    radius=1000, color='green', fill=True, fill_opacity=0.2,
    tooltip='Vùng kinh doanh'
).add_to(layer_vung)

folium.PolyLine(
    locations=[[10.74659, 106.69788],[10.72964, 106.71891]],
    color='blue',
    tooltip='Tuyến đường vận chuyển'
).add_to(layer_tuyen)
```

```python
folium.Marker(
    location=[10.74659, 106.69788],
    icon=folium.Icon(color='red', icon='home'),
    tooltip='Kho tổng'
).add_to(layer_diem)

folium.Marker(
    location=[10.72964, 106.71891],
    icon=folium.Icon(color='green', icon='shopping-cart'),
    tooltip='Vùng kinh doanh'
).add_to(layer_diem)
```

```python
layer_vung.add_to(m)
layer_tuyen.add_to(m)
layer_diem.add_to(m)
folium.LayerControl(position='topright', collapsed=False).add_to(m)
m
```
