```python
import folium
```

```python
m = folium.Map(location=[10.771421, 106.670282], zoom_start=13)
folium.Marker([10.771421, 106.670282], popup=f"Trung tâm phân phối").add_to(m)
```

```python
folium.Circle(location=[10.771421, 106.670282], radius=10000, color="green", fill=True, fill_opacity=0.1, popup="Vùng 3: Bán kính 10km").add_to(m)
folium.Circle(location=[10.771421, 106.670282], radius=5000, color="yellow", fill=True, fill_opacity=0.2, popup="Vùng 2: Bán kính 5km").add_to(m)
folium.Circle(location=[10.771421, 106.670282], radius=3000, color="red", fill=True, fill_opacity=0.3, popup="Vùng 1: Bán kính 3km").add_to(m)
m
```

1. Phân tầng theo từng khu vực
- Vùng đỏ (bán kính 3km): khoảng cách gần, phục vụ nhanh, phù hợp giao hàng hỏa tốc/ giao trong ngày.
- Vùng vàng (bán kính 5km): phục vụ trung bình, tốc độ giao hàng trung bình.
- Vùng xanh (bán kính 10km): khoảng cách xa, phục vụ xa, thời gian giao hàng lâu hơn.
2. Tối ưu chi phí logistic
- Khoảng cách càng xa, chi phí giao hàng càng cao (xăng, nhân công, thời gian)
- Cần giới hạn phạm vi giao nhanh, thu phí cao hơn đối với đơn hàng khoảng cách xa.
- Nếu nhiều khách cùng ở phạm vi xa, có thể cân nhắc mở thêm kho hàng ở khu vực đó.
