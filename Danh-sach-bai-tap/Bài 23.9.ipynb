```python
import folium
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
Quan_1 = np.random.normal(loc=[10.7769, 106.7009], scale=[0.015, 0.015])
Quan_3 = np.random.normal(loc=[10.7829, 106.6864], scale=[0.015, 0.015])
Quan_7 = np.random.normal(loc=[10.7296, 106.7218], scale=[0.015, 0.015])

location = np.vstack((Quan_1, Quan_3, Quan_7))
customer = pd.DataFrame(location, columns=['latitude', 'longitude'])
```

```python
kmean = KMeans(n_clusters=3, n_init=10, random_state=40)
customer['Quan'] = kmean.fit_predict(customer[['latitude', 'longitude']])
hub = kmean.cluster_centers_
```

```python
m = folium.Map(location=[customer['latitude'].mean(), customer['longitude'].mean()], zoom_start=10)
colors = ['blue', 'green', 'red']
for i, center in enumerate(hub):
  folium.Marker(
      location=[center[0], center[1]],
      tooltip=f'Hub {i+1}',
      icon=folium.Icon(color=colors[i])
  ).add_to(m)

for _, row in customer.iterrows():
  folium.CircleMarker(
      location=[row['latitude'], row['longitude']],
      radius=4,
      color=colors[int(row['Quan'])],
      fill=True
  ).add_to(m)

m
```

1. Tối ưu hóa chi phí

Thuật toán phân cụm tự động tìm ra "điểm trung tâm" của một nhóm khách hàng sao cho tổng khoảng cách từ điểm đó đến tất cả khách hàng trong nhóm là nhỏ nhất. Việc đặt trạm trung chuyển tại điểm trung tâm này sẽ giúp tối ưu hóa chặng giao hàng cuối, giảm thiểu chi phí vận hành và tiết kiệm đáng kể ngân sách cho doanh nghiệp.

2. Phân bổ nguồn lực

Sau khi phân cụm, ta thấy có những cụm rất nhỏ nhưng chứa mật độ khách hàng dày đặc (thường là ở trung tâm thành phố), và những cụm trải rộng nhưng rất ít khách hàng (ở ngoại ô). Với những nơi đông đúc, doanh nghiệp nên đầu tư kho có công suất xử lý lớn, trang bị nhiều phương tiện để giảm thời gian giao hàng. Ngược lại, đối với những cụm thưa thớt, nên sử dụng ít phương tiện giao hàng và đồng thời gom nhiều đơn đi giao cùng lúc để giảm chi phí giao hàng.
