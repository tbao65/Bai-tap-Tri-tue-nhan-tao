```python
import numpy as np
import pandas as pd
import folium
from sklearn.cluster import KMeans
```

```python
n = 150
segments = np.random.choice(['thấp', 'tầm trung', 'cao cấp'], size=n, p=[0.4, 0.4, 0.2])
average_price = []
for s in segments:
    if s == 'thấp':
        average_price.append(np.random.randint(15, 50))
    elif s == 'tầm trung':
        average_price.append(np.random.randint(70, 150))
    else:
        average_price.append(np.random.randint(300, 500))
average_price = np.array(average_price)
latitudes = np.random.normal(loc=10.776, scale=0.015, size=n)
longitudes = np.random.normal(loc=106.701, scale=0.015, size=n)
df = pd.DataFrame({
    'Latitude': latitudes,
    'Longitude': longitudes,
    'avg price': average_price,
})
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['ID_Cum'] = kmeans.fit_predict(df[['avg price']])

centroids = kmeans.cluster_centers_.flatten()
sorted_idx = np.argsort(centroids)

cluster_mapping = {
    sorted_idx[0]: ('Sinh viên', 'green'),
    sorted_idx[1]: ('Văn phòng', 'blue'),
    sorted_idx[2]: ('Sang trọng', 'red')
}

df['Phan_Khuc'] = df['ID_Cum'].apply(lambda x: cluster_mapping[x][0])
df['Mau_Sac'] = df['ID_Cum'].apply(lambda x: cluster_mapping[x][1])
```

```python
m = folium.Map(location=[10.776, 106.701], zoom_start=13)
for idx, row in df.iterrows():
  popup_info = f"<b>{row['Phan_Khuc']}</b><br>Giá trung bình quán: {row['avg price']:.1f}k"
  folium.Marker(
      location=[row['Latitude'], row['Longitude']],
      icon=folium.Icon(color=row['Mau_Sac']),
      tooltip=popup_info
  ).add_to(m)
m
```

**Mô tả**

Mô hình này nhằm phân cụm các quán ăn dựa trên mức giá đại diện (average price) bằng cách sử dụng thuật toán KMeans. Thông qua việc phân tích dữ liệu, mô hình tự động chia các quán thành các nhóm có đặc điểm giá tương đồng, từ đó hình thành các phân khúc như sinh viên, văn phòng và sang trọng.

Việc phân cụm giúp đơn giản hóa dữ liệu ban đầu, hỗ trợ việc nhận diện đặc điểm thị trường. Mô hình mang lại giá trị trong việc hỗ trợ người dùng lựa chọn quán ăn phù hợp với khả năng chi tiêu và vị trí địa lý.
